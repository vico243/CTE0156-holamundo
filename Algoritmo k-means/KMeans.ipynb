# ALGORITMO K-MEANS

#### Ejemplo del algoritmo de Machine Learning de aprendizaje no supervisado "K-means" para el dataset "Iris"

Primero se importa las librerias necesarias

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.patches as ptch

import sklearn
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

Se carga el dataset Iris

```python
Iris = datasets.load_iris()
print(dir(Iris))
```

Se imprime los datos de Iris

```python
print(Iris.data)
```

Se imprime los nombres de cada columna

```python
print(Iris.feature_names)
```

Se imprime las etiquetas de Iris

```python
print(Iris.target)
```

Se imprime los tipos de Iris

```python
print(Iris.target_names)
```

Se Convierte los Datos en Data Frames

```python
dkm = pd.DataFrame(Iris.data)
dkm.columns = ['Longitud_Sépalo','Ancho_Sépalo','Longitud_Pétalo','Ancho_Pétalo']
dkm.head()
```

```python
d = pd.DataFrame(Iris.target)
d.columns = ['et_real']
d.head()
```

Se agrupa los datos utilizando k-means

```python
agrupacion = KMeans(n_clusters=3, random_state=0)
print(agrupacion.fit(dkm))
```

Se imprime los datos de la agrupación

```python
print(agrupacion.labels_)
```

Se encuentra los centros de las agrupaciones

```python
c = agrupacion.cluster_centers_
print(c)
```

# Gráficos

Se muestra el gráfico de todos los datos con la agrupación real

```python
color = np.array(['coral','yellowgreen','violet'])
plt.figure(figsize=(15,5))

patch1 = ptch.Patch(color='coral', label='Setosa')
patch2 = ptch.Patch(color='yellowgreen', label='Versicolor')
patch3 = ptch.Patch(color='violet', label='Virginica')

plt.subplot(1,2,1)
plt.scatter(dkm.Longitud_Sépalo, dkm.Ancho_Sépalo, c=color[d.et_real],s=25)
plt.title('Clasificación Real en funcion del Sépalo')
plt.xlabel('Longitud del Sépalo')
plt.ylabel('Ancho del Sépalo')
plt.legend(handles=[patch1, patch2, patch3])

plt.subplot(1,2,2)
plt.scatter(dkm.Longitud_Pétalo, dkm.Ancho_Pétalo, c=color[d.et_real],s=25)
plt.title('Clasificación Real en funcion del Pétalo')
plt.xlabel('Longitud del Pétalo')
plt.ylabel('Ancho del Pétalo')
plt.legend(handles=[patch1, patch2, patch3])
```

Se grafica la clasificación real y la clasificación de K-means

```python
cc = np.array(['firebrick','darkolivegreen','m'])
plt.figure(figsize=(15,5))

patch4 = ptch.Patch(color='yellowgreen', label='Setosa')
patch5 = ptch.Patch(color='violet', label='Versicolor')
patch6 = ptch.Patch(color='coral', label='Virginica')

plt.subplot(1,2,1)
plt.scatter(dkm.Longitud_Pétalo, dkm.Ancho_Pétalo, c=color[d.et_real],s=25)
plt.title('Clasificación Real')
plt.xlabel('Longitud del Pétalo')
plt.ylabel('Ancho del Pétalo')
plt.legend(handles=[patch1, patch2, patch3])

plt.subplot(1,2,2)
plt.scatter(dkm.Longitud_Pétalo, dkm.Ancho_Pétalo, c=color[agrupacion.labels_],s=25)
plt.title('Clasificación con K-means')
plt.xlabel('Longitud del Pétalo')
plt.ylabel('Ancho del Pétalo')
plt.scatter(c[:, 2], c[:, 3], c=cc, s=100)
plt.legend(handles=[patch4, patch5, patch6])
```

Se reasigna las etiquetas a la clasificación y se grafica nuevamente

```python
et_kmeans = np.choose(agrupacion.labels_,[2,0,1]).astype(np.int64)
print(et_kmeans)
```

```python
ccr = np.array(['m','firebrick','darkolivegreen'])
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.scatter(dkm.Longitud_Pétalo, dkm.Ancho_Pétalo, c=color[d.et_real],s=25)
plt.title('Clasificación Real')
plt.xlabel('Longitud del Pétalo')
plt.ylabel('Ancho del Pétalo')
plt.legend(handles=[patch1, patch2, patch3])

plt.subplot(1,2,2)
plt.scatter(dkm.Longitud_Pétalo, dkm.Ancho_Pétalo, c=color[et_kmeans],s=25)
plt.scatter(c[:, 2], c[:, 3], c=ccr, s=100)
plt.title('Clasificación con K-means Reasignando las Etiquetas')
plt.xlabel('Longitud del Pétalo')
plt.ylabel('Ancho del Pétalo')
plt.legend(handles=[patch1, patch2, patch3])
```

# Análisis de Resultados

Se grafica la Matriz de Confusión

```python
titulo = [("Matriz de Confusión", None), ("Matriz de Confusión Normalizada", 'true')]

for t, normalizar in titulo:
    mc = confusion_matrix(d.et_real, et_kmeans, sample_weight= None, labels= None, normalize= normalizar)
    disp = ConfusionMatrixDisplay(confusion_matrix=mc, display_labels=Iris.target_names)
    disp = disp.plot(include_values= True, cmap= 'BuPu', ax=None, xticks_rotation=None)
    disp.ax_.set_title(t)

plt.show()
```

Se comprueba la exactitud

```python
print('Exactitud: ',"{0:.2f}".format(accuracy_score(d.et_real, et_kmeans)*100),'%')
```
